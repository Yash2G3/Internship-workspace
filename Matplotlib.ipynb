# 2.the object oriented plotting method 

```python
plt.figure()
# figure is just like a getting a canvas ready for painting 

axes=plt.axes([0.1,0.1,0.8,0.8]) #axis ka [left, bottom, width, height]➡️the axes variable that we have created is our canvas

axes.plot(x,y) # now go ahead do the same for putting in the data at the x and the y axi
axes.set_xlabel('The x axis')
axes.set_ylabel('The y axis')
axes.set_title('The title')
```

## how to create mutiple graph figures in object oriented plotting 

```python
fig=plt.figure()
# here the margin of both the figure objects is set set in such a way that they fit inside each other

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.3,0.4,0.5])   # the 0.2 sets the object inwards from the bigger one 
#                                       the 0.3b gives it a bit more heigth to get insde from the bottom too 
#                                       the 0.4 and 0.5 are smaller than the 0.8's so that the small graph stays inside
                                        
```

so basically [a,b,c,d] that we put in the add_axes method 

a= margin from left 

b= margin from bottom 

c= width of the graph 

d= heigth of the graph

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,2,2])
axes1.plot(x,y)
axes1.set_title('The biggest')
axes1.set_xlabel('the bottom x')
axes1.set_ylabel('the leftmost y')

axes2=fig.add_axes([0.235,0.18,1.75,1.75])
axes.plot(y,x)
axes2.set_title('the medium')
axes2.set_xlabel('the medium x')
axes2.set_ylabel('the medium y')


axes3=fig.add_axes([0.35,0.25,1.5,1.5])
axes3.plot(y,x)
axes3.set_title('THe smallest')
axes3.set_xlabel('the uppity x')
axes3.set_ylabel(['the inmost y'])
```

# subplotting axes or creating a table of axes

we make a table of axes in which each axes is a part of a list so we can easily grab them seperateyly

 tuple unpacking is used here to create subplots as each plot is assigned an index value when unpacked

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y,'r')
axes[0].set_title('the first')

axes[1].plot(y,x,'g')
axes[1].set_title('the second')
```

```python
axes
# see there are 2 axes created and stored in a list  we can acess each of them seperately
```

# figure size and DPI

figure size is used to change the sizes of the figure where we put in the axes 

the axes weill also be accomodated accordingly

syntax: figsize=(width,heigth, DPI='integer')
    
    DPI is dots per inch 

```python
fig,axes=plt.subplots(2,1,figsize=(10,6))#➡️2 rows and 1 column subplot with 2 axes fitting in figure of 10 ,6 width and heigth


axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
# ➡️to accomodate all the graphs with all the space
```

# how to set legend in a graph
legend is gives us info about  the final figure plotted on x and y
 
 for legend you need a label of the graph which is basically the name of the line plotted


```python
fig=plt.figure()

ax=fig.add_axes([0.1,0.1,0.8,0.8])

ax.plot(x,y,label='x and y')

ax.plot(x,x**3,label='x cubed')

ax.legend(loc=0)
# loc is used for choosing the best location for the legend to be placed loc=0 lets jup choose it automatically
```

## how to customize the plotted line 

color:'name pf your set color'
    
linewidth/lw = to chage the thickness of the plotted line

ls{linestyle} = can be set to '--' or '-.' or 'step' etc

alpha=to change the transparency of the line {takes in numerical values}

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green',linewidth=5,linestyle='--',alpha=0.2)
```

# markers and their customization 
used to mak the points on the line 

maker='type'➡️to select the type of marker that you want

markersize= integer value

markerfacecolor = 'color name'➡️to change the color of the marker

markerwidth = integer➡️to make the the boundaries more thick 

markeredgecolor = 'color name' ➡️to change the color of boudaries of the marker



```python
fig=plt.figure()

xs=fig.add_axes([0,0,1,1])

xs.plot(x,y,lw=2,marker='+',markersize=30)
```

# how to set limit to view data on both the axis

if we wanna view view data only in between 0 and 1 on the x axis 

and 0 to 5 on the y axis

syntax: 

for x limits➡️figure name.set_xlim([set of the starting values on the x axis])

for y limits➡️figure name.set_ylims([set of the starting and the ending points on the y axis])


```python
fig
```

```python
fig=plt.figure()

xs=fig.add_axes([0,0,1,1])

xs.plot(x,y,lw=2,marker='+',markersize=30)

xs.set_xlim([0,1]) 

# x limit set between 0 and 1

xs.set_ylim([0,5])

# y limit set between 0 and 5
```
